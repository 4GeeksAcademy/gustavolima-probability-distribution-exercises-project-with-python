**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution

def dbinom(x,size,prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.pmf(k=x,n=size,p=prob,loc=0)
    return result

def pbinom(q,size,prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    return result

def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    return result

def rbinom(n,size,prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.rvs(n=size,p=prob,size=n)
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.5) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

```python
# Probability functions, density and distribution for Normal Distribution
# Since Normal Dist is a continuous distribution, we can't calculate
# PMF, only PDF and CDF. 

def normalpdf(x, mean, std_dev):
    """
    Calculates the point estimate of the normal distribution
    """
    from scipy.stats import norm
    result = norm.pdf(x, mean, std_dev)
    return result

def normalcdf(x, mean, std_dev):
    """
    Calculates the cumulative of the normal distribution
    """
    from scipy.stats import norm
    result = norm.cdf(x, mean, std_dev)
    return result

def normalquant(p, mean, std_dev):
    """
    Calculates the quantile function from the normal distribution
    """
    from scipy.stats import norm
    result = norm.ppf(p, mean, std_dev)
    return result
```

```python
print(normalpdf(x = 12, mean = 50, std_dev = 54))
print(normalcdf(x = 12, mean = 50, std_dev = 54))
print(normalquant(p = 1, mean = 10, std_dev = 10))
```

```python
# Plot the Normal Distribution PDF

mean = 0
std_dev = 1
x = np.arange(-4, 4, 0.001)
# x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
pdfvalues = normalpdf(x, mean, std_dev)

plt.plot(x, pdfvalues, color='blue', label='pdf')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Gaussian Distribution')
plt.legend()
plt.show()
```

```python
# Plot the Normal Distribution PDF and CDF

mean = 0
std_dev = 1
x = np.arange(-4, 4, 0.001)
pdfvalues = normalpdf(x, mean, std_dev)
cdfvalues = normalcdf(x, mean, std_dev)

plt.plot(x, pdfvalues, color='blue', label='pdf')
plt.plot(x, cdfvalues, color='red', label='cdf')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Gaussian Distribution')
plt.legend()
plt.show()
```

```python
# Probability functions, density and distribution for Uniform Distribution
# Since Uniform Dist is a continuous distribution, we can't calculate
# PMF, only PDF and CDF. 

def unipdf(x, lowbound, highbound):
    """
    Calculates the point estimate of the uniform distribution
    """
    from scipy.stats import uniform
    result = uniform.pdf(x, lowbound, highbound)
    return result

def unicdf(x, lowbound, highbound):
    """
    Calculates the cumulative of the uniform distribution
    """
    from scipy.stats import uniform
    result = uniform.cdf(x, lowbound, highbound)
    return result

def unippf(x, lowbound, highbound):
    """
    Calculates the quantile function from the uniform distribution
    """
    from scipy.stats import uniform
    result = uniform.ppf(x, lowbound, highbound)
    return result
```

```python
print(unipdf(x = 15, lowbound = 10, highbound = 34))
print(unicdf(x = 15, lowbound = 10, highbound = 34))
print(unippf(x = 15, lowbound = 10, highbound = 34))
```
